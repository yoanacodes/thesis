```python
measure_name = 'Incidence'
cause_name = 'Diabetes mellitus type 2'
metric_name = 'Rate'

# subset = all_2018[(all_2018['measure_name'] == measure_name) & (all_2018['cause_name'] == cause_name) & (all_2018['metric_name'] == metric_name)]

# subset.head()
# # subset.shape

```

```python
def plot_data(measure_name, cause_name, metric_name, food):
    subset = all_2018[(all_2018['measure_name'] == measure_name) & (all_2018['cause_name'] == cause_name) & (all_2018['metric_name'] == metric_name)]
    
    food = food + '_median'
    
    fig, ax = plt.subplots()

    fig.set_figheight(8)
    fig.set_figwidth(16)
    
    x = subset['value']
    y = subset[food]
    
#     print('************************')
#     print(x, ':', y)
#     print('************************')
    
    
    ax.scatter(x, y, alpha=0.5)
    
    m, b = np.polyfit(x, y, 1)
    
    plt.plot(x, m*x+b)

    # for i, txt in enumerate(labels):
    # #     print(i, txt)
    #     ax.annotate(txt, ( np.stack(subset['val']).astype(None)[i], np.stack(subset[food]).astype(None)[i]))

    plt.xlabel("health - " + cause_name)
    plt.ylabel("food - " + food)
    plt.show()
    
    
```

```python
all_2018['cause_name'].unique()
```

```python
plot_data(measure_name, cause_name, metric_name, '15')
```

```python
foods = ['03', '07', '09', '13', '15', '16', '22', '27', '33', '34', '35', '37', '57']

for food in foods:
    plot_data(measure_name, cause_name, metric_name, food)
```

```python
nutrition_dictionary = pd.read_json('final_datasets/nutrition_filters.json')
nutrition_dictionary
```

```python
def plot_small_multiples(food, food_name):
    n_plots = all_2018['cause_name'].nunique()

    fig, axes = plt.subplots(nrows=38, ncols=2, figsize=(15, 2 * n_plots), tight_layout=True)
    axes_list = [item for sublist in axes for item in sublist] 

    # start with 05_median

    for disease, selection in all_2018.groupby("cause_name"):
        ax = axes_list.pop(0)
#         food = '05_median'

        x=selection['value']
        y=selection[food]

    #     ax.set_xlim(xmin=0)

        ax.scatter(x, y, alpha=0.6, s=10, label=disease)

        m, b = np.polyfit(x, y, 1)
        ax.plot(x, m*x+b)


        title = food_name + ' AND ' + disease
        ax.set_title(title)

        ax.set_xlabel(disease)
        ax.set_ylabel(food)


        ax.spines['left'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)


    for ax in axes_list:
        ax.remove()

    file_name = 'plots/' + food + '.pdf'
    fig.savefig(file_name)
    
    plt.close(fig)

#     plt.subplots_adjust(hspace=1)
```

```python
for nutrient in nutrition_dictionary.values:
#     print(nutrient[0], nutrient[2])
#     print(nutrient['label'], nutrient['value'])
    c_name = nutrient[2].replace('v', '') + '_median'
#     print(c_name)
    plot_small_multiples(c_name, nutrient[0])
    
```
